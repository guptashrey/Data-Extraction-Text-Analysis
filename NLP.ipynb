```python
# Import pandas
import pandas as pd

# Assign spreadsheet filename to `file`
file = 'cik_list.xlsx'
```

```python
# Load spreadsheet
xl = pd.ExcelFile(file)
```

```python
# Print the sheet names
print(xl.sheet_names)
```

```python
# Load a sheet into a DataFrame by name: df1
df1 = xl.parse('cik_list_ajay')
```

```python
import requests

print('Beginning file download with requests')

temp = 'https://www.sec.gov/Archives/'
for i in range(152):
    temp1 = temp + df1.iloc[i]['SECFNAME']
    url = temp1
    r = requests.get(url)
    fname = 'data/' + df1.iloc[i]['SECFNAME'].split('/')[3]
    with open(fname, 'wb') as f:
        f.write(r.content)
    
    print (fname)
```
